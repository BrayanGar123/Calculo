## Funcion Lineal
$y=3x+2$

3=pendiente

2= intercepto en y

$y=mx+b$



```python
(y2-y1)/(x2-x1)
```

```python
def funcLineal(x,m,b):
    return m*x+b

x=np.linspace(-4,4,100)
y=funcLineal(x,m=3,b=2)


fig, ax = plt.subplots()

# Graficar los datos
ax.plot(x, y)

# Mostrar solo el eje X e Y
ax.spines['top'].set_color('none')    # Ocultar borde superior
ax.spines['right'].set_color('none')  # Ocultar borde derecho

# Posicionar el eje X e Y en el centro
ax.spines['left'].set_position('zero')  # Mover el eje Y al centro
ax.spines['bottom'].set_position('zero')  # Mover el eje X al centro

# Ocultar las marcas de los ejes superior y derecho
ax.xaxis.set_ticks_position('bottom')  # Marcas solo en la parte inferior
ax.yaxis.set_ticks_position('left')    # Marcas solo en la parte izquierda


# Mostrar la gráfica
plt.show()
```

## Funcion constante

$y=c$

```python
def funConstante(x,c):
    return c*(x/x)


x=np.linspace(-4,4,100)
y=funConstante(x,c=3)

fig, ax = plt.subplots()

# Graficar los datos
ax.plot(x, y)

# Mostrar solo el eje X e Y
ax.spines['top'].set_color('none')    # Ocultar borde superior
ax.spines['right'].set_color('none')  # Ocultar borde derecho

# Posicionar el eje X e Y en el centro
ax.spines['left'].set_position('zero')  # Mover el eje Y al centro
ax.spines['bottom'].set_position('zero')  # Mover el eje X al centro

# Ocultar las marcas de los ejes superior y derecho
ax.xaxis.set_ticks_position('bottom')  # Marcas solo en la parte inferior
ax.yaxis.set_ticks_position('left')    # Marcas solo en la parte izquierda


# Mostrar la gráfica
plt.show()
```
